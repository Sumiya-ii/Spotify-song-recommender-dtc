```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import calplot
```

```python
# Loading the dataset
data = pd.read_csv("/Users/bayarmaaorsoo/Desktop/Breathe Mongolia/Data Analyses/Ulaanbaatar Data Analysis/Data CSV/tolgoit,-ulaanbaatar-air-quality.csv")
```

```python
data.head()
```

In this dataset, there is no raw data and every pollutants' data recorded in daily average AQI.

## Data Transformation

```python
#Doing some data cleaning for other data sets

#Converting data type of date (object to datetime)
data['date'] = pd.to_datetime(data['date'])

#MNB
data.columns = data.columns.str.replace(' ','')
data = data.sort_values(by = 'date', ascending = False)
cols_to_convert_data = data.columns.drop('date')
data[cols_to_convert_data] = data[cols_to_convert_data].apply(pd.to_numeric, errors='coerce')
```

## Data Visualizations

## Stacked bar chart - PM2.5 

```python
# Creating a new dataset for stacked bar chart
data1 = data
data1['year'] = data1['date'].dt.year
data1['month'] = data1['date'].dt.month
data1 = data1.bfill()
```

```python
column_names = ["2015", "2016", "2017", "2018", "2019", "2020", "2021"]

stack_bar_data = pd.DataFrame(columns = column_names)

# Percentage of 'Hazardous' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &(data1['pm25'] > 300)])) /
                           (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')

# Percentage of 'Very Unhealthy' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &
                                                              (data1['pm25'] <= 300) &
                                                              (data1['pm25'] > 200)])) /
                                            (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')

# Percentage of 'Unhealthy' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &
                                                              (data1['pm25'] <= 200) &
                                                              (data1['pm25'] > 150)])) /
                                            (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')

# Percentage of 'Unhealthy for sensitive peaople' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &
                                                              (data1['pm25'] <= 150) &
                                                              (data1['pm25'] > 100)])) /
                                            (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')

# Percentage of 'Moderate' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &
                                                              (data1['pm25'] <= 100) &
                                                              (data1['pm25'] > 50)])) /
                                            (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')


# Percentage of 'Good' days by years

for i in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    stack_bar_data = stack_bar_data.append({str(i): (len(data1[(data1['year'] == i) &
                                                              (data1['pm25'] <= 50) &
                                                              (data1['pm25'] >= 0)])) /
                                            (len(data1[data1['year'] == i]))}, ignore_index = True)

stack_bar_data = stack_bar_data.bfill().ffill()
stack_bar_data = stack_bar_data.drop_duplicates(keep='first')
```

```python
# Changing row names
stack_bar_data = stack_bar_data.set_axis(['Hazardous', 'Very Unhealthy', 'Unhealthy', 
             'Unhealthy for sensitive people', 'Moderate', 'Good'], axis=0)

# Transposing the dataset
stack_bar_data = stack_bar_data.T
```

```python
# Variables
labels = ['Hazardous', 'Very Unhealthy', 'Unhealthy', 'Unhealthy for sensitive people', 'Moderate', 'Good']
colors = ['darkred', 'purple', 'red', 'darkorange', 'yellow', 'green']
title = 'Percentage of Days by AQI Categories and Years\n'


def plot_stackedbar_p(df, labels, colors, title):
    fields = df.columns.tolist()
    
    # figure and axis
    fig, ax = plt.subplots(1, figsize=(20, 8))
# plot bars
    left = len(df) * [0]
    for idx, name in enumerate(fields):
        plt.barh(df.index, df[name], left = left, color=colors[idx])
        left = left + df[name]
# title and subtitle
    plt.title(title, loc='left', fontsize = 17)

# legend
    plt.legend(labels, bbox_to_anchor=([0.58, 1, 0, 0]), ncol=3, frameon=False)
# remove spines
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
# format x ticks
    xticks = np.arange(0,1.1,0.1)
    xlabels = ['{}%'.format(i) for i in np.arange(0,101,10)]
    plt.xticks(xticks, xlabels)
# adjust limits and draw grid lines
    plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
    ax.xaxis.grid(color='gray', linestyle='dashed')
plt.show()
    
plot_stackedbar_p(stack_bar_data, labels, colors, title)
```
